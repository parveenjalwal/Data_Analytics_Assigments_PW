```python
# 3. A Create a scatter plot using MatplotlibA
cars= sns.load_dataset('car_crashes')

plt.scatter(cars['total'],cars['speeding'],marker='*',color="orange")
plt.title("Car crashes")
plt.xlabel('total')
plt.ylabel('speeding')
```

```python
# 4. A How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?
tips = sns.load_dataset('tips')
tips_n = tips[['total_bill','tip','size']]
tips_n

tips_n.corr()
sns.heatmap(tips_n.corr(),cmap='coolwarm',annot=True)
```
